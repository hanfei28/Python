## Truth Value

* In Python, any object can be tested for Boolean truth value
* Any object cold be considered to be equivalent to Boolean true, unless this class defines 
 - a Bool method that returns false or  
 - has a len method that returns a zero length
 
 class MyClass:
     def __bool__(self):
         return False
         
     def __len__(self):
         return 0
 
The following objects are considered as false:
* False and None evalue to false
* Numeric Zero values: 0, 0.0, 0j, Decimal(0), fraction(0,x)
* Empty sequences/collections: '', ().[],{}
* Empty sets and ranges: set(), range(0)

Three Boolean operations:
1. and 
2. or 
3. not

## Strings vs Bytes

String and byte are not the same thing

* String in Python is a sequence of Unicode characters
* Bytes are a sequences of raw eight-bit values

```python
def main():
    b = bytes([0x41, 0x42, 0x43, 0x44])
    print(b)
    
    s = "This is a string."
    print(s)    
    
    #combine them
    #print(s+b) 
    print(s+b.decode("utf-8")) 
    print(s.encode("utf-8")+b) 
    print(s.encode("utf-32"))
    
    
if __name__ == "__main__":
    main()
```

## Template String

```python
from string import Template

def main():
    #using string formatting with format()
    s1 = "You are watching {0} by {1}".format("Advanced Python", "Fei Han")
    print(s1)
    
    #Create a template with placeholder
    templ = Template("You're reading ${title} by ${author}")
    
    # use the substitute method with keyword arguments
    s2 = templ.substitute(title="Python", author="Fei Han")
    print(s2)
    
    # use the substitute method with a dictionary
    record = {"author":"Fei Han", 
              "title":"Python"}
    s3=templ.substitute(record)
    print(s3)
    
if __name__ =="__main__":
    main()
```

* Template method is just text substitution
* with security concers, template method is a better way to go
